# The RAVDESS Dataset

RAVDESS is an acronym and stands for **Ryerson Audio-Visual Database of Emotional Speech and Song**. This is a large data set consisting of an audio and video database of approximately 24GB in size. However, since this project only has a demonstrative function, only a small part of it will be used to train the model faster and keep things simple. The dataset was downloaded from [Kaggle](https://www.kaggle.com/uwrfkaggler/ravdess-emotional-speech-audio) - a public web community that provides over 50,000 freely available datasets.

The portion of the dataset used here contains 1440 files: 60 recordings per actor x 24 actors, including 12 females and 12 males. Linguistic emotions include calm, happy, sad, angry, fearful, surprised, and disgusted expressions. The files are named according to a specific pattern which consists of seven parts in the following order:

- Modality (01 = full-AV, 02 = video-only, 03 = audio-only).
- Vocal channel (01 = speech, 02 = song).
- Emotion (01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised).
- Emotional intensity (01 = normal, 02 = strong). NOTE: There is no strong intensity for the 'neutral' emotion.
- Statement (01 = "Kids are talking by the door", 02 = "Dogs are sitting by the door").
- Repetition (01 = 1st repetition, 02 = 2nd repetition).
- Actor (01 to 24. Odd numbered actors are male, even numbered actors are female).

# Feature extraction

The pitch of a person's voice is a good indicator of their emotional state of mind. With sadness, for example, a person speaks with a lower pitch and rather monotonously, whereas with anger and joy the voice is higher and the speech melody more varied. Therefore, it makes sense to extract and utilize these features for emotion determination.

In machine learning and pattern recognition, a feature is a single measurable characteristic. Feature selection is a critical element of effective algorithms in pattern recognition, classification, and regression. Features are usually numeric, but structural features such as strings and graphs are used in syntactic pattern recognition. In speech recognition, phoneme recognition features may include noise ratios, length of sounds, relative power, filter matches, and many others.

As with any machine learning experiment, the first step is to collect data. The task is to extract features that can then be fed into an algorithm. Within this project, the following features are used:
- Mel spectograms
- Mel Frequency Cepstral Coefficients
- Chroma features

## The Mel Scale

In order to understand Mel spectrograms, we need to understand the Mel Scale first. Speech signals usually contain tones with different frequencies, where each tone has an actual frequency (Hz), and a subjective frequency. This means that the human ear perceives the tone frequency different based on the actual frequency of tones. It is more difficult for the ear to differentiate frequencies in high ranges than to differentiate frequencies that are in low ranges.

#### Example:
Consider the tones listed below with the frequencies 300Hz, 400Hz, 900Hz and 1000Hz: There is a 100Hz difference between the first two sounds as well as the last two sounds. However, the perceived distance between the first two sounds may seem greater than that between the last two, even though the actual difference is the same.

Basically, the Mel Scale (as seen below) is a logarithmic transformation of a signal's frequency which aims to represent this subjective perception of tone frequencies by the human ears.

```python
Image("data/Images/Mel-Skala.jpg", width=500, height=500)
```

For **frequencies up to about 500 Hz**, the frequency scale and the Mel Scale are almost proportional. Thus, a doubling of frequency from 100 Hz to 200 Hz results in a doubling of tonality from 100 mel to 200 mel (as perceived in the example of tones). For **frequencies greater than 500 Hz**, frequency and tonality have a nonlinear relationship. For example, the frequency of a tone must be increased from 1500 Hz to 10,000 Hz to achieve a doubling of tonality from 1100 mel to 2200 mel. I.e., in this frequency range, tone intervals are perceived to be smaller than they actually are.

The frequency f (hertz) can be converted to m (mel) as follows:

$ m = 2595 log_{10} (1 + \frac{f}{700}) $

## Mel Frequency Cepstral Coefficients (MFCC)

Sprachliche Signale enthalten in der Regel Töne mit unterschiedlichen Frequenzen, wobei jeder Ton eine tatsächliche Frequenz (Hz), und eine subjektive Frequenz hat. Das bedeutet, das menschliche Ohr nimmt die tatsächliche Frequenz unterschiedlich wahr. Für das Ohr ist es schwieriger, Frequenzen in hohen Bereichen zu differenzieren, als Frequenzen zu unterscheiden, die in niedrigen Frequenzbereichen liegen.

Die MFCC-Berechnung ist eine Nachbildung des menschlichen Gehörs, mit dem Ziel, das Funktionsprinzip des Ohrs künstlich zu implementieren. Mel Frequency Cepstral Coefficients beruhen auf der Unstimmigkeit zwischen den kritischen Bandbreiten des menschlichen Ohrs, also der Bandbreite, innerhalb dessen eine Tonwahrnehmung durch zusätzlichen Schall beeinflussbar und maskierbar ist. D.h. mit der MFCC-Berechnung wird versucht, die subjektive Wahrnehmung von Frequenzen durch das menschliche Ohr nachzuahmen. 




## Chroma Features

Chroma ist eine Darstellung, die zeigt, wie Menschen Farben zu Noten in Beziehung setzen. Die Chroma-Darstellung zeigt die Intensität jedes der 12 verschiedenen musikalischen Chromas der Oktave in jedem Zeitrahmen. Dieses Modul zur Merkmalsextraktion wird verwendet, um ein Chromagramm zu erstellen, das aus den zwölf Tonhöhenklassen aufgebaut ist. Die Variation dieser Tonhöhenklassen während des Audios ermöglicht es, 12 weitere Merkmale zu erzeugen, die zu den verschiedenen Tonhöhenklassen gehören und für die emotionale Sprachverarbeitung verwendet werden können. Die Tonhöhenklassen sind in der folgenden Reihenfolge: C, C#, D, D#, E, F, F#, G, G#, A, A#, B.

# Labels

Damit ein neuronales Netz Eingaben klassifizieren kann, benötigt es Wissen über die Klassen der Eingaben.  
Jedes Datenelement des Trainingsdatensatzes ist somit mit einem Label, seine Klasse, gekennzeichnet.
Während des Trainings des Netzes werden die Klassenlabel dazu verwendet, die Performance des Models zu beurteilen. Dazu wird die vom Model vorhergesagte Klasse des Datenelements mit der tatsächlichen Klasse verglichen. Diese Differenz bezeichnet man als Fehler, welcher während des Trainings minimiert werden soll.  
Nach dem Training, also bei Nutzung des Models, werden die Klassen verwendet, um die Eingaben, welche das Netzwerk noch nie zuvor verarbeitet hat, zu klassifizieren.  

*emotion_labels* sind hier alle Label, welche in dem Datensatz enthalten sind.   
Um die Performance des Netzwerks, in diesem prototypischen Aufbau, zu verbessern werden hier die Menge der Labels etwas eingeschränkt. Die hier verfügbaren Label sind jene aus dem Dictionary *available_emotions*

```python
emotion_labels = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}

available_emotions = {
    "angry",
    "sad",
    "disgust",
    "happy",
}
```

Um Audiosignale mittels einem Neuronalen Netz zu klassifizieren ist es notwendig im Vorfeld das Signal vorzuverarbeiten um besondere Merkmale, sogenannte Features, aus dem Signal zu extrahieren.  
Die ermittelten Features werden anschließend als Eingaben für das Model verwendet.  

Die gewünschten Merkmale, welche oben erleutert wurden, werden mit Hilfe der Bibliothek *librosa* berechnet und als Numpy-Array zurückgegeben.  
Damit erhält man für jedes Element des Datensatzes eine Sammlung von Features.

```python
def extract_audio_feature(X, sample_rate):
    result = np.array([])

    # mfcc
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0)
    result = np.hstack((result, mfccs))
        
    #chroma
    stft = np.abs(librosa.stft(X))
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T, axis=0)
    result = np.hstack((result, chroma))
        
    # mel
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T, axis=0)
    result = np.hstack((result, mel))
    return result
```

```python
def load_from_file(file_name):
    with soundfile.SoundFile(file_name) as sound_file:
        X = sound_file.read(dtype="float32")
        sample_rate = sound_file.samplerate
        return extract_audio_feature(X, sample_rate)
```

Die Funktion *load_training_data* dient dazu, die zum Training des Netzwerks benötigten Daten zu Laden und einen Train-Test-Split des Datensatzes zurück zu geben.  
Der Train-Split wird, wie der Name es vermuten lässt, rein zum Training des Models verwendet. Der Test-Split hingegen dient exklusiv zur Bewertung der Performance des Netzwerks. Hiermit kann gemessen werden, wie gut das Model die Trainingsdaten generalisiert hat und korrekte Vorhersagen auf unbekannten Daten liefert.   

```python
def load_training_data(test_size=0.2):
    X, y = [], []
    for file in glob.glob("data/Training_Data/Actor_*/*.wav"):
        basename = os.path.basename(file)
        emotion = emotion_labels[basename.split("-")[2]]
        if emotion not in available_emotions:
            continue
        features = load_from_file(file)
        X.append(features)
        y.append(emotion)
    return train_test_split(np.array(X), y, test_size=test_size, random_state=7)
```

```python
# load RAVDESS dataset, 75% training 25% testing
X_train, X_test, y_train, y_test = load_training_data(test_size=0.25)

print("[+] Number of training samples:", X_train.shape[0])
print("[+] Number of testing samples:", X_test.shape[0])
print("[+] Number of features:", X_train.shape[1])
```

Um das Model bequem zu erstellen, zu evaluieren, zu speichern und zu laden  wurden die Funktionen *create_model*, *evaluate_model*, *save_model* und *load_model* implementiert:  

* *create_model* erstellt ein neues Model, genauer ein Multi-Layer-Perzeptron-Classifier Netzwerk, welches anschließend mit den Trainingsdaten trainiert wird. 
* *evaluate_model* bietet die Möglichkeit das trainierte Model zu testen um dessen Performance zu ermitteln. Dieser Schritt ist allerdings optional, das Model verbessert sich nicht durch eine Evaluation. 
* *save_model* kann verwendet werden um das trainierte Model in eine Datei zu speichern, um es bei nächstem Programmstart erneut verwenden zu können. 
* *load_model* ist das passende Gegenstück zu *save_model*. Es dient dazu gespeicherte Modelle wieder zu laden. 

```python
def create_model():
    # best model, determined by a grid search
    model_params = {
        "alpha": 0.01,
        "batch_size": 32,
        "epsilon": 1e-08,
        "hidden_layer_sizes": (300,),
        "learning_rate": "adaptive",
        "max_iter": 500
    }
   
    # init multi layer perceptron classifier with best parameters (so far)
    model = MLPClassifier(**model_params)
    
    print("[*] Training the model...")
    model.fit(X_train, y_train)
    return model

def evaluate_model(model):
    # predict 25% of data to measure how good we are
    y_pred = model.predict(X_test)

    # calculate the accuracy
    accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
    print(f"Accuracy: {accuracy*100}%")

def save_model(model,filename):
    file = open(filename, 'wb')
    pickle.dump(model, file)

def load_model(filename):
    file = open(filename,'rb')
    return pickle.load(file)
```

```python
# train_model = True -> Neues Model wird erstellt und Trainiert.
# train_model = False -> Vorhandenes Model wird versucht aus Datei zu laden
train_model = False
model = None
if train_model: 
    model = create_model()
    evaluate_model(model)
    save_model(model, "model")
else:
    print("load model from disk")
    model = load_model("model")
    evaluate_model(model)
```

# Demonstration der emotionalen Spracherkennung

Im nachfolgenden soll die emotionale Spracherkennung, unter Nuztung des zuvor trainierten Models, nun demonstriert werden. Die Demonstration besteht dabei aus folgendem Szenario: Durch das Betätigen eines Start-Buttons beginnt das Szenario, wodurch in sieben-sekündigem Abstand verschiedene Bilder gezeigt werden. Gleichzeitig startet für jedes Bild eine im Hintergrund laufende seperate Sprachaufnahme, welche die gesprochene Reaktion des Probanden aufnehmen soll. Daher ist es äußerst wichtig, dass der Proband zu jedem Bild etwas sagt. Was gesagt wird ist hierbei völlig unbedeutend, da sich die Spracherkennung lediglich auf die übermittelte Emotion beschränkt, welche im Anschluss für jede Aufnahme, und mithilfe des trainierten Models, ermittelt wird. Abschließend werden die gezeigten Bildern mit ihren korrespondierenden Emotionen aufgezeigt was zu einer Diskussion über die Richtigkeit der erkannten Emotion und des Empfindens der Testperson während der Durchführung anregen soll.

### Welcher Zusammenhang besteht hierbei zur Sprachsteuerung?
Wie auch in der Ausarbeitung dieses Moduls beschrieben, gibt es viele Anwendungsbereiche, für die eine emotionale Sprachverarbeitung bzw. Sprachsteuerung denkbar wäre. Betrachtet man beispielsweise das Verhalten eines Sprachassistenten wie Amazon's Alex oder Apple's Siri wäre es durchaus denkbar, dass sich das Verhalten des Speakers an die erkannte Emotion des Bedieners anpasst. 

Beispiel: Der Anwender stellt eine Frage an den Sprachassistenten, welcher im Anschluss von diesem bearbeitet wird. Der Anwender erhält eine Anwort, welche seine Frage nicht zufriedenstellend beantwortet - weshalb er den Sprachassistenten genervt unterbricht und die Beantwortung seiner Frage terminiert. Der Speaker klassifiziert diese Emotion richtig und merkt sich, dass seine Antwort unzureichend war, wodurch er bei einer nächsten, ähnlichen Frage die Herangehensweise an die Beantwortung ändert. Wird festgestellt, dass der Anwender eine positive Emotion nachfolgend zur Antwort gibt, erkennt der Speaker, dass die verwendete Art und Weise der Bearbeitung zufriedenstellend war und nutzt diese Herangehensweise für ähnliche Situationen erneut.

Die Funktion *record* nimmt das Audiosignal auf, lässt die Features extrahieren und veranlasst das Model dazu eine Vorhersage zu treffen. Diese wird an den Aufrufer zurück gegeben.  
Die Aufnahmedauer beträgt hier 7 Sekunden, das für diesen Anwendungsfall der beste Kompromis zwischen "Es ist genung Zeit um etwas zu sagen" und "Es dauert zu lange bis ein neues Bild erscheint" ist. 

```python
def record():
    duration = 7 #Aufnahmedauer
    fs = 16000 #Samplerate
    recording = sd.rec(int(duration * fs),samplerate=fs, channels=1)
    sd.wait() #Aufnahme
    features = extract_audio_feature(recording.reshape(-1), fs)
    return model.predict(features.reshape(1, -1))[0]
```

Die in diesem Szenario verwendeten Testbilder befinden sich innerhalb des *data* Ordners. Deren Pfade und Kategorien (Korrespondierende erwartete Emotion) werden nun durch die Funktion *get_images* ermittelt und für die spätere Verwendung in einer Liste gespeichert und an den Aufrufer zurückgegeben.

```python
def get_images(path):
    """Returns all image paths and their corresponding emotion within the specified path"""
    images = []
    for root, directories, files in os.walk(traversal_path):
        for file in files:
            category = root.split("\\")[-1]  # folder name = emotion
            path = root + "/" + file
            image = {
                "path": path,
                "category": category
            }
            images.append(image)
    return images
```

Die hier programmierte Funktion *start_projection* kümmert sich um den korrekten Ablauf des Szenarios und wird beim Start aufgerufen. Zusätzlich ist sie für das korrekte Darstellen der Bilder zuständig und startet bei jedem neuen Bild eine neue Audioaufnahme, welche dem jeweiligen Bild mit der anschließend erkannten Emotion zugeordnet wird.

```python
traversal_path = "data/Images/Categories"
images = get_images(traversal_path)

def start_projection(button):
    for image in images:
        image_path = image.get("path")
        with output:
            clear_output()
            display(Image(image_path, width=600, height=600))
        detected_emotion = record()
        image["detected_emotion"] = detected_emotion
    clear_output()
```

Hier wird eine graphische Benutzeroberfläche bestehend aus einem Startbutton erzeugt, durch wessen Betätigung das Szenario gestartet werden kann.

```python
start_button = Button(
    description='',
    disabled=False,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Start projection',
    icon='play'
)
output = Output()
start_button.on_click(start_projection)
```
